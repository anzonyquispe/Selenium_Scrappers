```python
from datetime import datetime, timedelta

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

#Library
# ---
# For scrape
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

# Options driver
from selenium.webdriver.chrome.options import Options

# For dataframes
import pandas as pd
from pandas import json_normalize

# For simulate human behavior.
import time
from time import sleep
import random

# Clear data
import unidecode
#unaccented_string = unidecode.unidecode(accented_string)

# Manipulate Json files
import json
import re
import numpy as np
from datetime import datetime
from selenium.webdriver.common.keys import Keys
```

```python
# Setting
options= Options()
options.add_argument("--start-maximized")
driver = webdriver.Chrome(ChromeDriverManager().install(),options=options)
# we open our page
driver.get("https://www.sbs.gob.pe/app/pp/EstadisticasSAEEPortal/Paginas/TIActivaTipoCreditoEmpresa.aspx?tip=B#")
```

```python
# Get data from dates
data_date = pd.DataFrame( {'date_real' : pd.date_range(start="01-01-2000",end="08-31-2022") })

data_date['date_info'] = data_date.date_real.dt.strftime("%d/%m/%Y")
data_split = data_date.date_info.str.split("/", expand= True ).astype(int)
data_split.columns = ["Day", "Month", "Year"]
data_date = pd.concat([ data_date, data_split ], axis = 1 )

data_date['date_real'] = pd.to_datetime(data_date.date_info, format="%d/%m/%Y")

# Get the last 5 days per month
final_date_data = data_date.sort_values('date_real').groupby([ "Year" , "Month" ]).tail(1).reset_index(drop = True)

dates_list = final_date_data.date_real.tolist()
```

```python
# Get data from dates
data_date = pd.DataFrame( {'date_real' : pd.date_range(start="03-31-2001",end="08-31-2022", freq = "D") })

data_date['date_info'] = data_date.date_real.dt.strftime("%d/%m/%Y")
data_split = data_date.date_info.str.split("/", expand= True ).astype(int)
data_split.columns = ["Day", "Month", "Year"]
data_date = pd.concat([ data_date, data_split ], axis = 1 )

data_date['date_real'] = pd.to_datetime(data_date.date_info, format="%d/%m/%Y")

# Get the last 5 days per month
final_date_data = data_date.sort_values('date_real').groupby([ "Year" , "Month" ]).tail(1).reset_index(drop = True)

dates_list = final_date_data.date_real.tolist()
```

```python
# Lists of information
list_dates = []
list_table = []

# Loop of dates
for date1 in dates_list:
    
    # Iterator
    i = 1
    
    # loop of information
    while i < 8:
        
        driver.switch_to.default_content()
        # Get information of dates
        day_float = date1.day
        month_float = date1.month
        year_float = date1.year
        
        # Input date of information
        date_string = date1.strftime("%d/%m/%Y")
        print("\n",date_string)
        driver.execute_script( f"document.getElementById('ctl00_cphContent_rdpDate_dateInput').value = '{date_string}'")
#         driver.find_element_by_id("ctl00_cphContent_rdpDate_dateInput_wrapper").send_keys(date_string)
        driver.find_element_by_id("ctl00_cphContent_rdpDate_dateInput").send_keys(Keys.RETURN)
        time.sleep( 3 )
        
        table_clean1 = 15
        table_clean2 = 15
        table_clean3 = 15
            
        # Try getting 1 table in another frame
        try:
            # Get tables
            section_id_table = driver.find_element_by_id( "ctl00_cphContent_rpgActualMn_OT" )
            table_html = section_id_table.get_attribute('outerHTML')
            table_pd = pd.read_html( table_html )

            # generate table objects
            table0, table1 = table_pd[0], table_pd[1]
            table_clean = table1.copy()

            # Number of rows of second table
            tb1_rows = table1.shape[0]
            drop_rows = np.arange( 1, tb1_rows + 2 )

            # Clean first table
            table0 = table0.drop( drop_rows, axis = 0 )

            # Assign values to new table
            col_name = table0.iloc[0, 0]
            col_values = table0.iloc[1:, 0].tolist()
            table_clean.insert( loc = 0, column= col_name, value = col_values)
            table_clean1 = table_clean.copy()
            
            
        except:
            print( "No 1 table in the same frame" )
            table_clean1 = "Null"
            
            
        # Try getting 1 table in another frame
        try:
            # Switch elements
            driver.switch_to.frame(driver.find_element_by_id("ctl00_cphContent_pnContenidoAnteriorMnASP"))
            section_tab_table = driver.find_element_by_xpath( "/html/body/form/table/tbody" )
            table_html = section_tab_table.get_attribute('outerHTML')
            table_pd = pd.read_html( table_html )

            # Cleaning tablr
            table_clean2 = table_pd[1].iloc[2:, :].copy().reset_index( drop = True )

            # Changing switch
            driver.switch_to.default_content()
            
        except:
            print( "No 1 table in another frame" )
            table_clean2 = "Null"
            
            
        # Try getting table of 6 layers in another frame
        try:
            
            # Switch elements
            driver.switch_to.frame(driver.find_element_by_id("ctl00_cphContent_pnContenidoAnteriorMnASP"))
            section_tab_table = driver.find_element_by_xpath( "/html/body/form/table/tbody" )
            table_html = section_tab_table.get_attribute('outerHTML')
            table_pd = pd.read_html( table_html )

            # Cleaning table
            table_clean3 = pd.concat([ table_pd[3], table_pd[6].iloc[:, 1:] ], axis = 1 )

            # Changing switch
            driver.switch_to.default_content()
        
        except:
            print( "No 2 tables in another frame" )
            table_clean3 = "Null"
            
            
        # Check what to save
        if ( ( isinstance( table_clean3, str ) == False ) and ( table_clean3.shape[1] > 5 ) ):
            
            # Data with no frame
            table_clean3.to_excel( fr"B_RawData/sbs/B_RawData/bancos/table_clean_{day_float}_{month_float}_{year_float}.xlsx", 
                                  index = False, header= False )
            # To break
            i = 8
            list_dates.append( date1 )
            list_table.append( "Table" )
            print( "tabl3 Yes")
            
        elif ( ( isinstance( table_clean2, str ) == False ) and ( table_clean2.shape[1] > 5 ) ):
            
            # Data with frame and 2 tables
            table_clean2.to_excel( fr"B_RawData/sbs/B_RawData/bancos/table_clean_{day_float}_{month_float}_{year_float}.xlsx", 
                                  index = False, header= False )
            # To break
            i = 8
            list_dates.append( date1 )
            list_table.append( "Table" )
            print( "tabl2 Yes")
            
        elif ( ( isinstance( table_clean1, str ) == False ) and ( table_clean1.shape[1] > 5 ) ):
            # Data with frame and 1 table
            table_clean1.to_excel( fr"B_RawData/sbs/B_RawData/bancos/table_clean_{day_float}_{month_float}_{year_float}.xlsx", 
                               index = False )
            
            # To break
            i = 8
            list_dates.append( date1 )
            list_table.append( "Table" )
            print( "tabl1 Yes")
            
        # No table information
        # Go to a previous day
        else:
            # Day ago
            print( f"GO {i} before")
            date1 = date1 - timedelta(days= 1)
            print( date1 )
            i = i + 1
            list_dates.append( date1 )
            list_table.append( "No Table" )
      
    
```
